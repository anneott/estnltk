```python
import estnltk
from estnltk import Text
```

### Basic use case

```python
#Load the tagger

from estnltk.taggers.estner.ner_tagger import NerTagger

nertagger = NerTagger()
```

```python
#Create a text object
text = Text('''Eesti Vabariik on riik Põhja-Euroopas. Eesti piirneb põhjas üle Soome lahe Soome Vabariigiga.''')
```

```python
#Ner layer will be added to the text
nertagger.tag(text)
```

```python
#Spans in the ner layer consist of two attributes: text with words that make up the named entity and nertag that shows
#the category of the named entity: "LOC", "PER" or "ORG"
text.ner
```

For some use cases it might be better to have the output layer with a tag for each word. This can be used for visualizing or making manual changes to the layer.

```python
from estnltk.taggers.estner.word_level_ner_tagger import NerTagger as WordLevelNerTagger

word_level_ner = WordLevelNerTagger()
```

```python
word_level_ner.tag(text)
```
